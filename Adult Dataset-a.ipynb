# Lendo arquivo e preparando para inclose5_CARM

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
import plot_classes_dataset as plt_dtset

def save_sorted_dtset(path, dt_name, r_ord, labels, save=False):
    dt = plt_dtset.Dataset(path, dt_name, r_ord, labels)
    ready_matrix = np.loadtxt(dt.path_filename)
    sorted_dt = dt.sort_matrix(ready_matrix, 'asc')
    if save:
        np.savetxt(path+dt_name+'_asc', sorted_dt, fmt='%u', delimiter=' ', newline='\n')
    return sorted_dt

h = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week','native-country', 'class' ]
data = pd.read_csv("/home/julia/Downloads/adult.data", names=h).replace(' ?', None)
data
```

```python
test = pd.read_csv("/home/julia/Downloads/adult.test", names=h).replace(' ?', None)
test
```
