```python
idx_labels_controle = dt.index[dt['GroupAED (1 - controles, 2 - responsivos, 3 - resistentes)'] == 1]
```

```python
z_scores = np.array([])
for column in dt[qntf]:
    stand_d = dt[column].loc[idx_labels_controle].dropna().std(ddof=0)
    #print(especFarm[column].loc[idx_labels_controle])
    m = dt[column].loc[idx_labels_controle].dropna().mean()
    print(column, 'mean = ', m, 'stand_d = ', stand_d)
    if stand_d != 0:
        z_score = ((dt[column].dropna() - m)/stand_d)
        
        z_score = z_score.dropna()
        
        for i, z in z_score.iteritems():
            if z < 2 and z > -2:
                dt.loc[i,column] = 0
            elif z >= 2:
                dt.loc[i, column] = 1
            else:
                dt.loc[i, column] = -1
        
        
    else:
        dt[column] = 0
```

```python
discretize_cols = ['SideHS (1 - controles, 2 - MRI-negativa, 3 - AHD, 4 - AHL)', 
                   'Regimen (0 - sem DAEs, 1 - uma DAES, 3 - 2 ou mais DAEs) ', 
                   'ipsiGLU', 'contraGLU']
binary_cols = ['Depressão (0 - não, 1 - sim)','Tratamento_Antidepressivo (0 - não, 1 - sim)',
                             'BLOQ_SODIO_RAP', 'BLOQ_SOD_LENTO','POTEN_GABA', 'PROT_SIN', 'ANTI-GLU', 'ipsiINS',
                             'ipsiGPCt', 'contraINS', 'contraGPCt']

new_data = np.array([])
new_cols2 = []
n_rows = len(df)
for column in df.iloc[:,:-1]:
    if column in binary_cols:
        col = df[column].replace(9999, 0).to_numpy(dtype='int').reshape((n_rows, 1))
        #print('col = ', df[column].to_numpy(dtype='int').reshape((n_rows, 1)))
        new_data = np.concatenate((new_data, col), axis=1)
        new_cols2 += [column]
        
    else:
        uniques = df[column].unique().astype('int')
        print(uniques)
        uniques = np.delete(uniques, np.where(uniques == 9999))
        uniques = np.delete(uniques, np.where(uniques == 0))
            

        print(column, ' = ', len(uniques), uniques)
        aux = np.zeros((n_rows,len(uniques)), dtype=int)
        for i, value in df[column].iteritems():
            for j in range(len(uniques)):
                if value == uniques[j]:
                    aux[i,j] = 1
                else:
                    aux[i,j] = 0

        if new_data.size:
            new_data = np.concatenate((new_data, aux), axis=1)
        else:
            new_data = aux


        for j in uniques:
            new_cols2 += [column+str(j)]
                
    

        
print(new_data.shape)
print(new_cols2)
```
